匀加速直线运动：

$y = v_0 t - \frac{1}{2} g t^2$


```python
v0 = 5      # initial velocity
g = 9.81    # acceleration
t = 1.0     # time

y = v0*t - 0.5*g*t**2

print('y = ', y)
```

计算：$ Angle = \tan^{-1} (y/x)$, 结果用“角度”表示。

```python
import math

x = 10.0   # horizontal position
y = 10.0   # vertical position

arcangle = math.atan(y/x)

angle = (arcangle/math.pi)*180

print ('angle = ', angle)
```

在同一张图里画$f(t) = t^2$, 和$g(t) = e^t$，t的变化区域为[-2,2]

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-2, 2, 100)

f_values = t**2
g_values = np.exp(t)

plt.plot(t, f_values, 'r', t, g_values, 'b--')
plt.xlabel('t')
plt.ylabel('f and g')
plt.legend(['t^2', 'e^t']) #图例
plt.title('Plotting of two functions (t^2 and e^t)')
plt.grid('on')
plt.axis([-3, 3, -1, 10])  #t取值区域为-3到3，y取值区域为-1到10
plt.show()
```

用subplot命令同时画多张图
---

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

plt.subplot(2, 1, 1)   #两行，1列，其中第一张图
v0 = 5
g = 9.81
t = np.linspace(0, 1, 11) #11个点，包括首尾，10等分
y = v0*t - 0.5*g*t**2

plt.plot(t, y, '*')
plt.xlabel('t (s)')
plt.ylabel('y (m)')
plt.title('Ball moving vertically')

plt.subplot(2, 1, 2)  #第二个图
t = np.linspace(-2, 2, 100)
f_values = t**2
g_values = np.exp(t)

plt.plot(t, f_values, 'r', t, g_values, 'b--')
plt.xlabel('t')
plt.ylabel('f and g')
plt.legend(['t^2', 'e^t'])
plt.title('Plotting of two functions (t^2 and e^t)')
plt.grid('on')
plt.axis([-3, 3, -1, 10])

plt.tight_layout()
plt.savefig('sample_plot.png')
plt.show()
```

格式化输出，$g(t) = t \sin(t)$

```python
from math import sin

t0 = 2
dt = 0.55

t = t0 + 0*dt; g = t*sin(t)
print('{:6.2f} {:8.3f}'.format(t, g))

t = t0 + 1*dt; g = t*sin(t)
print('{:6.2f} {:8.3f}'.format(t, g))

t = t0 + 2*dt; g = t*sin(t)
print('{:6.2f} {:8.3f}'.format(t, g))

t = t0 + 3*dt; g = t*sin(t)
print('{:6.2f} {:8.3f}'.format(t, g))
```
