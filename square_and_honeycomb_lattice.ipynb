```python
%matplotlib inline
from pylab import *

import matplotlib.pyplot as plt
```

$\epsilon(k_x, k_y)= t \left( cos k_x + cos k_y \right)$

```python
ttt = 2.0
def tba2df(x, y):
    return ttt*(cos(x)+cos(y))
```

```python
x = linspace(-pi, pi, 100)
y = linspace(-pi, pi, 100)
X,Y = meshgrid(x, y)
Z = tba2df(X, Y).T
```

```python
fig, ax = plt.subplots()

p = ax.pcolor(X/(2*pi), Y/(2*pi), Z, cmap=cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max())
cb = fig.colorbar(p, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(6,6))

cnt = ax.contour(Z, cmap=cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
```

```python
fig, ax = plt.subplots()

im = ax.imshow(Z, cmap=cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
im.set_interpolation('bilinear')

cb = fig.colorbar(im, ax=ax)
```

$\epsilon(k_x, k_y) = \cos(k_x) + \cos(\frac{k_x + \sqrt{3}y}{2}) + cos(\frac{x - \sqrt{3}y}{2})$

```python
alpha = 0.5*sqrt(3)
alpha
```

```python
def honeycomb(x,y):
    return cos(x) + cos(0.5*x + alpha*y) + cos(0.5*x - alpha*y)
```

```python
x = linspace(-3*pi, 3*pi, 300)
y = linspace(-3*pi, 3*pi, 300)
X,Y = meshgrid(x, y)
Z = honeycomb(X, Y).T
```

```python
fig, ax = plt.subplots()

p = ax.pcolor(X/(2*pi), Y/(2*pi), Z, cmap=cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max())
cb = fig.colorbar(p, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(6,6))

cnt = ax.contour(Z, cmap=cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
```

```python
fig, ax = plt.subplots()

im = ax.imshow(Z, cmap=cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
im.set_interpolation('bilinear')

cb = fig.colorbar(im, ax=ax)
```
