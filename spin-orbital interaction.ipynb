# “自旋-轨道”耦合

在量子力学的发展史上，当人们试图写出电子运动所需满足的波动方程的时候，狭义相对论已经提出了，因此任何理论需要符合相对论协变性，或与狭义相对论不冲突，是写出正确的电子方程的必要条件。

但这些努力都不成功。比较典型的是“克莱因-戈登方程”：

$$ - \hbar^2 \frac{\partial^2 }{\partial t^2} \psi = \left( - \hbar^2 c^2 \nabla^2 + m^2 c^4 \right)\psi $$

这个方程其实就是把相对论的“能量-动量公式”

$$ E^2 = p^2 c^2 + m^2 c^4 $$

里的E和p分别用能量算符和动量算符替代得到的。但如果把这个方程看做是电子运动需满足的波动方程，则会碰到如下两个困难：（1）负几率问题；（2）负能量问题。不过后来这个方程在量子场论兴起后，获得了新的意义，被认为是描述标量场（一个分量）的方程。

在量子力学中取得成功的第一个方程反而是一个“非相对论”性的波动方程——薛定谔方程，它算出了氢原子的玻尔能级。

$$ i \hbar \frac{\partial }{\partial t} \psi = \left( - \frac{\hbar^2}{2m}\nabla^2 + V \right)\psi $$

但用薛定谔方程来研究原子物理的问题的时候，还是会觉得不完美，比如相对论效应如何考虑，比如自旋如何理解等。

原子物理的问题显然与电磁学相关，其核心问题是电子在原子核产生的电磁场中运动，而狭义相对论也是电磁理论发展的自然结果（电磁学是否满足伽利略协变性），这提示我们原子物理，与电磁学，及狭义相对论有很紧密的联系。

根据电磁学，我们除了可以用电磁（E），磁场（B）来描述物理问题外，我们还可以引入A（磁矢势），φ（电标势）来描述，为了符合相对论协变性，我们把它们统一写为Aμ=(A0, A1, A2, A3)。

在这种记号下电子在原子核中的势能V就是eA0，并且我们需要在做正则量子化的时候，把电磁场的动量也考虑进去。这样电子在原子中的哈密顿量就可以写为：

$$ H = \frac{(\hat p - e \vec A)^2}{2m} + eA_0 $$

对应波动方程是：

$$ i \hbar \frac{\partial }{\partial t} \psi = \left[ \frac{(\hat p - e \vec A)^2}{2m} + eA_0  \right]\psi$$

这个波动方程是考虑了电磁场之后电子满足的薛定谔方程。假设Aμ=0，就是非相对论的自由电子满足的波动方程。

在非相对论量子力学中，我们是通过实验直接引入自旋概念的，并认为自旋角动量与磁矩之间的关系是：

$$ \vec \mu_s = - g_s \frac{e}{2m} \vec S $$

这里gs是自旋的朗德因子，与电子作轨道运动导致的朗德因子gl=1不同，这里gs=2。

自旋在磁场B中的能量是塞曼能，对应

$$H_z =  - \mu \cdot B \\= - g_s \frac{e}{2m}\frac{\hbar}{2} \sigma \cdot B \\ = \frac{e\hbar}{2m} \sigma \cdot B $$

（注：这里形式上e取正。）

这里引入了泡利矩阵σ来描述自旋S，

$$ \sigma_x = \pmatrix{ 0 & 1 \cr 1 & 0   }, \sigma_y = \pmatrix{ 0 & - i \cr i & 0 }, \sigma_z = \pmatrix{ 1 & 0 \cr 0 & -1 } $$

考虑了自旋后，自由电子在非相对论情形下的哈密度算符是，

$$H = \frac{\left( \sigma \cdot p \right)\left( \sigma \cdot p \right)}{2m} = \pmatrix{ p^2/2m & 0 \cr 0 & p^2/2m }  $$

考虑电磁场后，

$$ \frac{1}{2m} \sigma \cdot \left( \hat p - eA  \right) \sigma \cdot \left( \hat p - eA  \right)  $$

在这个计算中考虑到A与动量算符p不对易，化简可得：

$$ \frac{1}{2m}\left( \hat p - eA \right)^2 - \frac{e \hbar}{2m} \sigma \cdot B $$

这里利用了电磁学里磁场与磁矢势A之间的关系：

$$ B = \nabla \times A $$

这里，我们由更基础的物理原理推出电子的朗德因子gs是2。

进一步，考虑相对论性的能量-动量关系，我们将其改写为能够描述自旋1/2的形式：

$$ \left( \frac{E}{c} - \sigma \cdot p \right) \left( \frac{E}{c} + \sigma \cdot p \right) = (mc)^2 $$

改写成算符的形式：

$$ \left(i \hbar \frac{\partial }{\partial x_0} + i \hbar \sigma \cdot \nabla   \right) \left(i \hbar \frac{\partial }{\partial x_0} - i \hbar \sigma \cdot \nabla   \right) \phi = (mc)^2 \phi $$

这里x0=ct，φ表示两分量的波函数。

假设

$$ \phi^L = \phi \\ \phi^R = \frac{1}{mc} \left(i \hbar \frac{\partial }{\partial x_0} - i \hbar \sigma \cdot \nabla   \right) \phi$$

这意味着：

$$ \left[ i \hbar \sigma \cdot \nabla - i \hbar \partial_0 \right] \phi^L = - mc \phi^R \\\left[ - i \hbar \sigma \cdot \nabla - i \hbar \partial_0 \right] \phi^R = - mc \phi^L $$

以上两式相加，相减：

$$ - i \hbar (\sigma \cdot \nabla) \left( \phi^R - \phi^L \right) -  i \hbar \partial_0 \left( \phi^L + \phi^R \right) = - mc \left( \phi^L + \phi^R \right)\\  i \hbar (\sigma \cdot \nabla) \left( \phi^L + \phi^R \right) + i \hbar \partial_0 \left( \phi^R - \phi^L \right) = - mc \left( \phi^R - \phi^L \right)  $$

在此基础上定义一个新的四分量波函数ψ，

$$\psi = \pmatrix{\psi_A \cr \psi_B } = \pmatrix{ \phi^R + \phi^L \cr \phi^R - \phi^L }$$

满足，

$$ \pmatrix{-i \hbar \partial_0 & - i \hbar \sigma \cdot \nabla \cr i \hbar \sigma \cdot \nabla & i \hbar \partial_0   }\pmatrix{\psi_A \cr \psi_B } = - mc \pmatrix{\psi_A \cr \psi_B } $$

这就是自由电子的狄拉克方程，引入γ矩阵：

$$\gamma_0 = \pmatrix{ I & 0 \cr 0 & -I  }, \gamma_k = \pmatrix{ 0 & \sigma_k \cr - \sigma_k & 0 } $$

改写成常见的形式：

$$ \left(i \hbar \gamma_\mu \frac{\partial }{\partial x_\mu} - mc \right)\psi = 0 $$

考虑不含时的量子力学问题，得到定态狄拉克方程，写成常见的形式：

$$ \left( c \alpha \cdot p + \beta m c^2 + V  \right)\psi = E \psi $$

这里α和β是两个4乘4的矩阵，

$$ \alpha = \pmatrix{ 0 & \sigma \cr \sigma & 0 }, \beta = \pmatrix{ 1 & 0 \cr 0 & -1 } $$

考虑电子在电磁场中运动，

$$ \left[ \sigma \cdot \left( p - e A \right)  \right]\psi_B = \frac{1}{c} \left(E- eA_0 - mc^2 \right)\psi_A \\ - \left[ \sigma \cdot \left( p - e A \right)  \right]\psi_A = - \frac{1}{c} \left( E - eA_0 + mc^2 \right)\psi_B $$

这里eA0相当于前式中的V（电子在电场中的静电势能），上式的特点是两份量波函数ψA和ψB是混合在一起的，单独某一个无法归一化。

我们可以在形式上把ψB消掉，

$$\left[ \sigma \cdot \left( p - e A \right)  \right] \left[ \frac{c^2}{E-eA_0 + mc^2} \right] \left[ \sigma \cdot \left( p - e A \right)  \right]  \psi_A \\=  \left[ E- eA_0 - mc^2  \right] \psi_A $$ 

我们现在的任务是在非相对论极限下，把狄拉克方程化简，比如在非相对论情形下，我们说的能量其实是相对论的能量减去电子的静能量，并且等于动能加势能。

$$ \tilde{E} = E - mc^2 \approx \frac{mv^2}{2}+V \\= \frac{mv^2}{2}+ e A_0  $$

这意味着：

$$ \frac{c^2}{E-eA_0 + mc^2} = \frac{1}{2m}\left[1 - \frac{\tilde{E} - eA_0}{2 mc^2  }  \right] \\ = \frac{1}{2m} \left[ 1- \frac{v^2}{4c^2} \right] $$

忽略(v/c)平方项，得到：

$$ \frac{1}{2m} \sigma \cdot \left( p - eA \right)  \sigma \cdot \left( p - eA \right) \psi_A = \left( \tilde{E} - eA_0 \right) \psi_A $$

上式整理简化后是：

$$ \left[ \frac{1}{2m} \left( p - eA \right)^2 + eA_0 - \frac{e \hbar}{2m }\sigma \cdot B \right]\psi_A = \tilde{E} \psi_A$$

这里就只剩下一个二分量的波函数ψA了，同时上式就是考虑电磁场后的定态薛定谔方程。

即使是在非相对论极限下（A=0，E=mc^2），仍然有一部分ψA变成了ψB，

$$ \psi_B \approx \frac{\sigma \cdot p}{2mc} \psi_A$$

归一条件：

$$ \int d^3 x \left( \psi^\dagger_A \psi_A +  \psi_A^\dagger \left(  \frac{p^2}{4 m^2 c^2 } \right)  \psi_A \right) = 1 $$

引入新的两份量波函数Ψ，

$$ \Psi =\left( 1 + \frac{p^2}{8 m^2 c^2} \right)\psi_A  = \Omega \psi_A$$

使得：

$$ \int d^3 x \Psi^\dagger \Psi = \int d^3 x \psi_A^\dagger \left[ 1 + \frac{p^2}{4 m^2 c^2}  \right] \psi_A = 1$$

考虑磁矢势A=0，但电标势A0不为0，ψA满足的方程是：

$$H_A \psi_A \\ = \left[ \left( \sigma \cdot p \right) \frac{1}{2m}\left(1-  \frac{\tilde{E} - eA_0}{2m c^2} \right) \left( \sigma \cdot p \right) + eA_0 \right] \psi_A \\ = \tilde{E} \psi_A  $$

在上式中我们考虑了相对论(v/c)^2项的修正，但ψA的问题是其有一部分会变为ψB，为了避免这个问题，我们考虑Ψ，这里面是包含了ψB的成分的，并且已经在形式上归一化，其满足的波动方程是：

$$\Omega^{-1} H_A \Omega^{-1}\Psi  = \tilde{E} \Omega^{-2}\Psi $$

我们现在把上式展开，并忽略掉其中比(v/c)^2更高级的贡献，得到：

$$ \left[ \frac{p^2}{2m} + eA_0 - \frac{p^4}{8 m^3 c^2} - \frac{e \hbar }{4 m^2 c^2}\sigma \cdot \left( \vec E \times \hat p \right) - \frac{e \hbar^2}{ 8 m^2 c^2} \nabla \cdot \vec E  \right] \Psi = \tilde{E}\Psi$$

由于在开始的时候假设A=0，所以上式中没有出现塞曼项，但这里出现了自旋轨道耦合项（上式中的第四项），上式中的第三项可以从相对论能量动量公式的展开中直接看出来，最后一项是达尔文项，与原子中的电荷分布有关。

现在重点看一下自旋轨道耦合项，

$$ - \frac{e \hbar }{4 m^2 c^2}\sigma \cdot \left( \vec E \times \hat p \right) $$

上式中的E表示电场强度，

$$ \vec E = - \nabla A_0 = -\frac{1}{e} \nabla \left( eA_0 \right) = - \frac{1}{e}\nabla V \\ = -\frac{1}{e}\hat e_r \frac{\partial V}{\partial r} = -\frac{1}{e} \frac{ \vec r}{r} \frac{\partial V}{\partial r}$$

代回自旋轨道耦合项：

$$ \frac{\hbar}{4 m^2 c^2 } \left( \frac{1}{r}\frac{\partial V}{\partial r} \right) \sigma \cdot  (\vec r \times \hat p ) \\ = \frac{1}{2m^2 c^2}\left( \frac{1}{r} \frac{\partial V}{\partial r} \right)  S \cdot L $$

这就是常见的自旋轨道耦合项。

小结一下，狄拉克方程是一个四分量的波动方程，在非相对论极限下，狄拉克方程退化为一个二分量的波动方程，对应包含塞曼能的薛定谔方程，并且会自然地得到电子的自旋朗德因子是2，在保留相对论效应(v/c)平方项的近似下，我们得到了自旋轨道耦合项。
